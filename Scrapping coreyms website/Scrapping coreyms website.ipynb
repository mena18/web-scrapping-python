# Scrapping coreyms

### the goal is to get all avaliables videos from coreyms website and save tehm in csv file 
### so scrapping all pages not only the first page

```python
from bs4 import BeautifulSoup 
import requests
import csv
import time
```

```python
def pr(soup,csv_writer):
    articles = soup.find_all('article')
    counter=0

    for article in articles:
        try:
            #rint()
            text = article.h2.a.text
            src = article.iframe['src'].split('?')[0].split('/')[-1]
            src = f'https://www.youtube.com/watch?v={src}'
            csv_writer.writerow([text,src])
            counter+=1


        except:
            pass
    return counter
```

```python
csv_file = open('scrapping_coreyms.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'video_link'])



i=1
while True:

    page = requests.get(f'https://coreyms.com/page/{i}').text
    soup = BeautifulSoup(page,'lxml')
    print(f"page {i}")
    print('-'*50)
    counter=pr(soup,csv_writer)
    if(counter==0):
        break;
    print("number of articles",counter)
    time.sleep(0.5)
    print('-'*50,end="\n\n")
    i+=1
    
        
print("Done")
csv_file.close()
```
